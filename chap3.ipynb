### RUN AS INTENDED (DO NOT CHANGE ANYTHING.)

```python
""" Initialise and Establish Dataset """
import pandas as pd

search_terms = product

search_terms = pd.read_pickle("support/_current_/searchTerms.pkl")
youtube_comment_list = pd.read_pickle("support/%s/youtube/comment_list.pkl" % search_terms)
reddit_comment_list = pd.read_pickle("support/%s/reddit/comment_list.pkl" % search_terms)
pc_gamer = pd.read_pickle("support/%s/pc_gamer/pc_gamer.pkl" % search_terms)
product_desc = pd.read_pickle("support/%s/product_desc/product_desc.pkl" % search_terms)
user_manual = pd.read_pickle("support/%s/user_manual/user_manual.pkl" % search_terms)
```

```python
""" Join all comments into a corpus """
def combine_text(list_of_text):                     # define combine_text to take (list_of_text)
    combined_text = ' '.join(list_of_text)          # do this
    return combined_text                            # and give combined_text back

print("Joining Youtube comments...")
all_youtube_comments = combine_text(youtube_comment_list)
print(all_youtube_comments)
print()
print("Joining Reddit comments...")
all_reddit_comments = combine_text(reddit_comment_list)
print(all_reddit_comments)
```

```python
""" Remove stopwords """
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
""" Remove stopwords """
all_youtube_comments = " ".join(word for word in all_youtube_comments.split() if word not in stopwords)
all_reddit_comments = " ".join(word for word in all_reddit_comments.split() if word not in stopwords)
all_pc_gamer = " ".join(word for word in pc_gamer.split() if word not in stopwords)
all_product_desc = " ".join(word for word in product_desc.split() if word not in stopwords)
all_user_manual = " ".join(word for word in user_manual.split() if word not in stopwords)

print("Split Youtube comments:")
print(all_youtube_comments)
print()
print("Split Reddit comments:")
print(all_reddit_comments)
print()
print("Split PC Gamer review:")
print(all_pc_gamer)
print()
print("Split Product desc:")
print(all_product_desc)
print()
print("Split User manual:")
print(all_user_manual)
```
